```python
import pandas as pd
import sqlite3
import os
```

```python
# Load the Excel file
# Correct file path with extension
file_path = r'C:\Users\info\OneDrive\Desktop\ABC Project\PharmData.xlsx'

# Load the Excel file
data = pd.read_excel(file_path, sheet_name='Data')

# Connect to SQLite database (or create one)
db_path = 'pharm_data.db'  # Update with your desired database path
conn = sqlite3.connect(db_path)

# Create and populate dimension tables
def create_and_populate_dimension(table_name, columns, conn, data):
    # Select unique records for the dimension table
    dimension_data = data[columns].drop_duplicates()
    dimension_data.to_sql(table_name, conn, if_exists='replace', index=False)

# Distributor Dimension Table
create_and_populate_dimension("dim_distributor", ["Distributor"], conn, data)

# Customer Dimension Table
create_and_populate_dimension(
    "dim_customer", ["Customer Name", "City", "Country", "Latitude", "Longitude"], conn, data
)

# Product Dimension Table
create_and_populate_dimension("dim_product", ["Product Name", "Product Class"], conn, data)

# Sales Team Dimension Table
create_and_populate_dimension(
    "dim_sales_team", ["Name of Sales Rep", "Manager", "Sales Team"], conn, data
)

# Create and populate fact table
fact_table_columns = ["Distributor", "Customer Name", "Product Name", "Quantity", 
                      "Price", "Sales", "Month", "Year", "Name of Sales Rep"]
fact_table = data[fact_table_columns]
fact_table.to_sql("fact_sales", conn, if_exists='replace', index=False)

# Verify the created tables
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Created tables:")
print(tables)

# Close the connection
conn.close()
```

```python
# Connect to the SQLite database
db_path = 'pharm_data.db'  # Path to your SQLite database
conn = sqlite3.connect(db_path)

# Get the list of all tables in the database
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)

# Display top 5 rows from each table
for table_name in tables['name']:
    print(f"Top 5 rows from table: {table_name}")
    query = f"SELECT * FROM {table_name} LIMIT 5;"
    table_data = pd.read_sql_query(query, conn)
    print(table_data)
    print("\n")

# Close the connection
conn.close()
```
